#### Ahora haremos una vista general de las temporadas de la juventus

```python
#crearemos el dataframe "Juventus" en donde almacenaremos todas las temporadas en las que este fue campeon que es desde el
# 2011 hasta el 2019
df_juventus = [df11, df12, df13, df14, df15, df16, df17, df18,]
teamJ = 'Juventus Juventus'
df_pts = []

for df in df_juventus:
    df_filtrado = df.loc[df['position'] ==1]
    df_pts.append(df_filtrado)
    
df_J_final = pd.concat(df_pts, ignore_index=True)

#ahora crearemos una lista con las temporadas para anadirlas al dataframe
num_temporadas = len(df_J_final)
temporadas = []
inicio = 2011  

for i in range(num_temporadas):
    temporada = f"{inicio + i}/{(inicio + i + 1) % 100:02d}"
    temporadas.append(temporada)

df_J_final['Season'] = temporadas

df_J_final
```

De aqui podemos observar como su temporada con mas puntos fue la 2013/14 en la que logro hacer 102pts ya que fue la temporada en la que gano mas partidos, sin embargo la temporada en la que no perdio ningun partido fue la 2011/12, mientras que su temporada con menos puntos fue la 2011/12, esto lo vamos a visualizar mejor con los siguientes graficos.

```python
def grafico(dfi, figname, Eje_X, Eje_Y, Title_X, Title_Y, tinta): #Funcion para crear graficos 
    df = dfi.copy()
    df = df[[Eje_X, Eje_Y]].groupby(Eje_X).sum()
    df = df.reset_index()
    plt.figure(figsize=(15,5))
    plt.plot(df[Eje_X], df[Eje_Y], marker='o', color=tinta)
    plt.title(figname)
    plt.xlabel(Title_X)
    plt.ylabel(Title_Y)
    
    plt.grid(True)
    plt.xticks(df[Eje_X], rotation=45)
    
    
    plt.show()
    
    return 

grafico(df_J_final, 'Puntos Por Temporada', 'Season', 'Pt', 'Temporada', 'Puntos', 'green')

```

## Analisis General 2009-2019

```python
#Concatenamos todos los dataframes

dfs = [df09, df10, df11, df12, df13, df14, df15, df16, df17, df18]
df_combined = pd.concat(dfs, ignore_index=True)

numeric_columns = ['Pt', 'Goals made', 'Goals suffered', 'Won', 'Net', 'lose']
for col in numeric_columns:
    df_combined[col] = pd.to_numeric(df_combined[col], errors='coerce')

#Con la tabla normal quedan equipos debido a un error de tipos de datos, entonces las siguientes lineas eliminan esos equipos
df_combined['team'] = df_combined['team'].astype(str)
invalid_teams = df_combined[df_combined['team'].str.contains('-\d')]['team'].unique()

df_combined = df_combined[~df_combined['team'].str.contains('-\d')]

# Sumar los valores de cada equipo
df_total = df_combined.groupby('team')[numeric_columns].sum().reset_index()


df_total = df_total.sort_values(by='Pt', ascending=False)
df_total
```

```python
#Media, Maximo y minimo de 10 las 10 temporadas
df_total.describe()
```

### Goles anotados Por temporada

```python
season_names = ['2009/10', '2010/11', '2011/12', '2012/13', '2013/14', '2014/15', '2015/16', '2016/17', '2017/18', '2018/19']
dfs = [df09, df10, df11, df12, df13, df14, df15, df16, df17, df18]

# Añadir la columna de temporada a cada dataframe
for i, df in enumerate(dfs):
    df['season'] = season_names[i]

df_combined = pd.concat(dfs, ignore_index=True)
df_combined['season'] = df_combined['season'].astype(str)
df_goles_por_temporada = df_combined.groupby('season')['Goals made'].sum().reset_index()

df_goles_por_temporada
```

```python
def grafico_bar(dfi, figname, Eje_X, Eje_Y, Title_X, Title_Y, tinta):
    df = dfi.copy()
    df = df[[Eje_X, Eje_Y]].groupby(Eje_X).sum()
    df = df.reset_index()
    df = df.sort_values(by=Eje_Y, ascending=True)
    
    plt.bar(df[Eje_X], df[Eje_Y], width=0.6, color=tinta)
    plt.title(figname)
    plt.xlabel(Title_X)
    plt.ylabel(Title_Y)   
    plt.xticks(df[Eje_X], rotation=45)

    plt.show()
    
    return

grafico_bar(df_goles_por_temporada, 'Goles por Temporada', 'season', 'Goals made', 'Temporada', 'Goles', 'Purple')
```

### Titulos Obtenidos

```python
df_max_puntos = [df09, df10, df11, df12, df13, df14, df15, df16, df17, df18,]
df_pts = []

for df in df_max_puntos:
    df_filtrado = df.loc[df['position'] ==1]
    df_pts.append(df_filtrado)
    
df_max_puntos = pd.concat(df_pts, ignore_index=True)

#ahora crearemos una lista con las temporadas para anadirlas al dataframe


df_max_puntos = df_max_puntos.groupby('team')['position'].sum().reset_index()
df_max_puntos
```
